### Задача 1.2

*Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома\
Если реновации дома не было, то в новом признаке поставьте 0*

```python
df['delta_renovated'] = (df['yr_renovated'] - df['yr_built'])
df['delta_renovated'] = df['delta_renovated'].fillna(0)
df = df.loc[df['delta_renovated'] >= 0]

df.head(10)
```

### Задача 1.3

*Создайте признаки года продажи, месяца продажи*

```python
df['date'] = pd.to_datetime(df['date'])

df['sale_year'] = df['date'].dt.year
df['sale_month'] = df['date'].dt.month

df.head(10)
```

### Задача 1.4

```python
df = df.drop(columns=['date', 'zipcode', 'lat', 'long'])
df.head(10)
```
